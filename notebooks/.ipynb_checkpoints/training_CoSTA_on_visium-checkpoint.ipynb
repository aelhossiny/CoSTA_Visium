# Introduction

* This projects aims to apply [CoSTA](https://bmcbioinformatics.biomedcentral.com/articles/10.1186/s12859-021-04314-1) pipeline on Visium datasets. Also modify CoSTA architecture to include transformer layers.
* Here we've used the 10X Genomics Visium 4 human prostate sample downloaded from here [10X Genomics Datasets](https://www.10xgenomics.com/resources/datasets?query=&page=1&configure%5Bfacets%5D%5B0%5D=chemistryVersionAndThroughput&configure%5Bfacets%5D%5B1%5D=pipeline.version&configure%5BhitsPerPage%5D=500&menu%5Bproducts.name%5D=Spatial%20Gene%20Expression).
* Samples were initially analyzed using Seurat V4 spaital data analysis pipeline [Here](https://www.10xgenomics.com/resources/datasets?query=&page=1&configure%5Bfacets%5D%5B0%5D=chemistryVersionAndThroughput&configure%5Bfacets%5D%5B1%5D=pipeline.version&configure%5BhitsPerPage%5D=500&menu%5Bproducts.name%5D=Spatial%20Gene%20Expression). `creating_gene_exp_images.Rmd` notebook was used to generate the spatial gene expression images from the samples. Briefly, each dataset was imported using `Load10X_Spatial` function, then normalized using `SCTransform()` function, then filtered for top highly variable 3000 genes, then gene expression images (200*200) were generated using  `SpatialFeaturePlot()`. 
* Images were processed to (85*85) grayscale images to make computational faster, and all images were concatenated in (3000,1,85,85) array.

# Analyzing Acinar Cell Carcinoma Sample

## Loading the processed images array

Here We're loading the spatial gene expression arrays generated in the `converting_images_to_np_array` notebook. We'll be using the 85*85 images array. Then we're plotting the frequency of the mean pixel intensity of the genes images to filter out images with mostly background. We're setting the cut-off to 20. Then we're normalizng the pixels intensity values across all images as described in the paper.

```python
h,w = 85,85
dataset = np.load("../results/HVGeneExpArrays_85/acinar_cell_carcinoma.npy")
genes = listdir("../results/HVGeneExpImgs/acinar_cell_carcinoma/")
genes = [re.sub('.jpeg$', '', gene) for gene in genes]
dataset = dataset.reshape(dataset.shape[0],1,h,w)
```

```python
n, bins, patches = plt.hist(x= np.mean(dataset.reshape(dataset.shape[0],85*85), axis =1), bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean Pixel Value Freq')
maxfreq = n.max()
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```

```python
fil = np.where(dataset.mean(axis = (1,2,3)) > 20)[0].tolist()
dataset = dataset[fil]
genes = np.array(genes)[fil].tolist()
```

```python
dataset = utils.normalize(expr= dataset, genes = genes, h = h, w = w)
```

## Loading model

Here we're using the same model described in the paper. We only changed the stride of the first layer to 1 as this showed superior performance

```python
net = model.ConvNet85()
net.apply(utils.weights_init)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
net
```

## Gene pseudolabels before training

```python
##visualization without training
import umap.umap_ as umap
X_all_tensor = torch.tensor(dataset).float()
y_pred = net.forward_feature(X_all_tensor)
y_pred = torch.Tensor.cpu(y_pred).detach().numpy()
au_tar, y_label, embedding = utils.evaluation(y_pred,n_neighbors=5,min_dist=0.0,
                                        num_cluster=10,cluster_method='GMM') 
original = y_label.copy()
embedding = umap.UMAP(n_neighbors=5, min_dist=0.0, n_components=2,
                      metric='correlation').fit_transform(y_pred)
```

```python
embedding_df = pd.DataFrame(embedding)
embedding_df.columns=['UMAP1','UMAP2']
embedding_df["Proton"]=original
```

```python
plt.figure(figsize=(10,10))
f=sns.lmplot(x='UMAP1', y='UMAP2',data=embedding_df,hue="Proton",
             fit_reg=False,legend=False,scatter_kws={'s':20})
f.fig.suptitle('Initial Clusters before training', weight='bold').set_fontsize('13')
#plt.title("Initial Clusters before training")
plt.savefig("../results/figures/Initial_clusters_acinar_cell_carci.png", format = 'png', bbox_inches="tight", dpi = 200)
```

```python
# Interacitve UMAP
import umap.umap_ as umap
embedding = umap.UMAP(n_neighbors=15, min_dist=0.3, n_components=2,
                      metric='correlation').fit(y_pred)
import umap.plot
hover_data = pd.DataFrame(
{'cluster':np.argmax(au_tar, axis = 1),
 'Gene': genes,
 'Index': range(dataset.shape[0])})
p = umap.plot.interactive(embedding, labels= np.argmax(au_tar, axis = 1), hover_data=hover_data, point_size=5,color_key_cmap='Paired', background='black')
umap.plot.show(p)
```

## Training Model

```python
y_pred_acinar, au_tar_acinar, NMI_history_acinar = train.train_costa(net = net, expr_array = dataset, num_epoch = 50, batch_size = 256, t1 = 0.8, t2 = 1.2, cluster_method= 'GMM')
```

```python
plt.plot(NMI_history_acinar)
plt.xlabel('epoch')
plt.ylabel('NMI')
plt.savefig("../results/figures/NMI_acinar_cell_car.png", format = 'png', bbox_inches="tight", dpi = 200)
```

We can see that NMI increases with epoches 

## Genes pseudolabels after training

```python
import umap.umap_ as umap
embedding_after_acinar = umap.UMAP(n_neighbors=5, min_dist=0.0, n_components=2,
                  metric='correlation').fit_transform(y_pred_acinar)
```

```python
embedding_after_acinar_df = pd.DataFrame(embedding_after_acinar)
embedding_after_acinar_df.columns=['UMAP1','UMAP2']
embedding_after_acinar_df["Proton"]=original
embedding_after_acinar_df["new"]=np.argmax(au_tar_acinar, axis = 1)
```

```python
f=sns.lmplot(x='UMAP1', y='UMAP2',data=embedding_after_acinar_df,hue="new",
             fit_reg=False,legend=False,scatter_kws={'s':20})
f.fig.suptitle('Clusters after training', weight='bold').set_fontsize('13')
#plt.title("Clusters after training")
plt.savefig("../results/figures/clusters_after_training_acinar_cell_carci.png",format = 'png', bbox_inches="tight", dpi = 200)
```

```python
f=sns.lmplot(x='UMAP1', y='UMAP2',data=embedding_after_acinar_df,hue="Proton",
             fit_reg=False,legend=False,scatter_kws={'s':20})
f.fig.suptitle('Clusters after training (original labels)', weight='bold',y=1.05 ).set_fontsize('13')
#plt.title("Clusters after training (with original labels)")
plt.savefig("../results/figures/clusters_after_training_acinar_cell_carci_w_orig_labels.png",format = 'png', bbox_inches="tight", dpi = 200)
```

**We can clearly see that the genes gets re-clustering to different positions (or neighborhoods) upon training forming new clusters**

```python
# interactive UMAP
import umap.umap_ as umap
embedding_after_acinar = umap.UMAP(n_neighbors=5, min_dist=0.0, n_components=2,
                      metric='correlation').fit(y_pred_acinar)
import umap.plot
hover_data = pd.DataFrame(
{'cluster':np.argmax(au_tar_acinar, axis = 1),
 'Gene': genes,
 'Index': range(dataset.shape[0])})
p = umap.plot.interactive(embedding_after_acinar, labels= np.argmax(au_tar_acinar, axis = 1), hover_data=hover_data, point_size=5,color_key_cmap='Paired', background='black')
umap.plot.show(p)
```
